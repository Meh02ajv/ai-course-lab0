Exercise 5: Pandas Data Analysis
Analyze a dataset of student performance.

```python
import pandas as pd
import numpy as np


np.random.seed(42)
n_students = 200

data = {
    "student_id": range(1000, 1000 + n_students),
    "major": np.random.choice(["CS", "Math", "Physics", "Biology"], n_students),
    "year": np.random.choice([1, 2, 3, 4], n_students),
    "exam_score": np.random.normal(75, 10, n_students).clip(0, 100),
    "assignments_completed": np.random.randint(0, 11, n_students),
    "hours_studied": np.random.normal(15, 5, n_students).clip(1, 40),
}

df = pd.DataFrame(data)

# Introduce some NaN values
df.loc[np.random.choice(n_students, 10), "exam_score"] = np.nan
df.loc[np.random.choice(n_students, 5), "hours_studied"] = np.nan


# Task 1: Data Cleaning and Exploration

# Display basic information about the dataset
print(df.info())

# Count missing values
print(df.isna().sum())

# Fill missing exam_score with mean
df["exam_score"].fillna(df["exam_score"].mean())

# Fill missing hours_studied with median
df["hours_studied"].fillna(df["hours_studied"].median())
```

```python
import pandas as pd
import numpy as np

# Task 2: Analysis

# Calculate and display the average exam_score by major
df.groupby("major")["exam_score"].agg(["mean"])

# Find the major with the highest average exam_score
df.groupby("major")["exam_score"].agg(["mean"]).max()

# Calculate the correlation between hours_studied and exam_score
df["hours_studied"].corr(df["exam_score"])

# Create a new column 'performance' with categories
df.loc[df["exam_score"] >= 90, "performance"] = "Excellent"
df.loc[(df["exam_score"] >= 80) & (df["exam_score"] < 90), "performance"] = "good"
df.loc[(df["exam_score"] >= 70) & (df["exam_score"] < 80), "performance"] = "Average"
df.loc[(df["exam_score"] < 70), "performance"] = "Needs Improvement"
```

```python
# Task 3: Advanced Analysis

# For each major and year combination, calculate

#Number of students
df.groupby(['major','year']).count()

#Average exam score
df.groupby(['major','year'])['exam_score'].agg(['mean'])

#Average hours studied
df.groupby(['major','year'])['hours_studied'].agg(['mean'])

#Identify top 5 students based on exam_score
top_students = df.nlargest(5, 'exam_score')


# Create a pivot table showing average exam_score by major (rows) and year (columns)
pivot = df.pivot_table(values='exam_score',index='major',columns='year',aggfunc='mean')




```

Exercise 6: Data Visualization
Create meaningful visualizations using the dataset from Exercise 5.

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Task 1: Distribution Visualization

# Create a figure with 2 subplots side by side

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

#Left: Histogram of exam scores with KDE overlay
sns.histplot(df["exam_score"], bins=20, kde=True, ax=axes[0], color="skyblue", edgecolor="black")
axes[0].set_title("Distribution of Exam Scores", fontsize=14, fontweight="bold")
axes[0].set_xlabel("Exam Score")
axes[0].set_ylabel("Frequency")

#Right: Box plot of exam scores by major
sns.boxplot(x="major", y="exam_score", data=df, ax=axes[1], palette="Set2")
axes[1].set_title("Exam Scores by Major", fontsize=14, fontweight="bold")
axes[1].set_xlabel("Major")
axes[1].set_ylabel("Exam Score")

plt.tight_layout()
plt.show()
```

```python
# Task 2: Relationship Visualization

# Create a scatter plot of hours_studied vs exam_score
sns.lmplot(data=df,x="hours_studied",y="exam_score",hue="major",scatter_kws={"alpha": 0.6},line_kws={"color": "black"})
plt.title("Hours Studied vs Exam Score by Major")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Task 3: Advanced Dashboard
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 1. Bar chart: Average exam score by major
means = df.groupby("major")["exam_score"].mean()
axes[0, 0].bar(means.index, means.values, color=["red", "blue", "green", "orange"])
axes[0, 0].set_xlabel("Major")
axes[0, 0].set_ylabel("Mean Exam Score")
axes[0, 0].set_title("Average Exam Score by Major")

# 2. Count plot: Number of students by year
sns.countplot(data=df, x="year", ax=axes[0, 1], palette="pastel")
axes[0, 1].set_title("Number of Students by Year")
axes[0, 1].set_xlabel("Year")
axes[0, 1].set_ylabel("Count")

# 3. Heat map: Correlation matrix of numerical columns
numeric_df = df.select_dtypes(include=["number"])
correlation = numeric_df.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f", ax=axes[1, 0])
axes[1, 0].set_title("Correlation Matrix of Numerical Columns")

# 4. Violin plot: Exam score distribution by performance category
sns.violinplot(data=df,x="performance",y="exam_score",inner="quartile",palette="Set2",ax=axes[1, 1])
axes[1, 1].set_title("Exam Score Distribution by Performance Category")
axes[1, 1].set_xlabel("Performance")
axes[1, 1].set_ylabel("Exam Score")


plt.tight_layout()
plt.show()
```

Exercise 7: Integration Challenge
Combine NumPy, Pandas, and Matplotlib to solve a mini data science problem.

```python
np.random.seed(42)
n_customers = 500

# Generate customer data
ages = np.random.randint(18, 70, n_customers)
income = np.random.normal(50000, 20000, n_customers).clip(15000, 150000)
purchase_freq = np.random.poisson(5, n_customers)
avg_purchase_value = np.random.normal(100, 30, n_customers).clip(10, 500)

# Create DataFrame
customers = pd.DataFrame(
    {
        "age": ages,
        "income": income,
        "purchase_frequency": purchase_freq,
        "avg_purchase_value": avg_purchase_value,
    }
)

# Calculate customer lifetime value (CLV)
max_frequency = customers["purchase_frequency"].max()

customers["churn_risk"] = 1 - (customers["purchase_frequency"] / max_frequency)

customers["CLV"] = (customers["purchase_frequency"]* customers["avg_purchase_value"]* (1 + customers["churn_risk"]))




# Create age groups: 18-25, 26-35, 36-50, 51-70
customers.loc[(customers["age"] >= 18) & (customers["age"] <= 25), "age_group"] = ("18-25")

customers.loc[(customers["age"] >= 26) & (customers["age"] <= 35), "age_group"] = (
    "26-35"
)
customers.loc[(customers["age"] >= 36) & (customers["age"] <= 50), "age_group"] = (
    "36-50"
)
customers.loc[(customers["age"] >= 51) & (customers["age"] <= 70), "age_group"] = (
    "51-70"
)


# For each age group, calculate:
#Number of customers
selected = customers["age_group"].agg(["count"])
print(selected)

#Average income
k=customers.groupby("age_group")["income"].agg(["mean"])
print(k)

#Average CLV
p=customers.groupby("age_group")["CLV"].mean()
print(p)

#Total CLV
o=customers.groupby("age_group")["CLV"].agg(["sum"])
print(o)


# Identify top 10% of customers by CLV



# Create visualizations:

#Scatter plot of income vs CLV 
sns.scatterplot(data=customers,x="income",y="CLV",hue="age_group",)
plt.title("Income vs CLV by age")
plt.xlabel("Income")
plt.ylabel("CLV")
plt.legend(title="age_group()")
plt.show()

#Bar chart of average CLV by age group
plt.bar(p.index, p.values, color="skyblue")
plt.xlabel("Age Group")
plt.ylabel("Average CLV")
plt.title("Average Customer Lifetime Value by Age Group")
plt.show()

#Correlation heatmap
numeric_d = customers.select_dtypes(include=["number"])
correlation = numeric_d.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")


```

BRIEF ANALYSIS PARAGRAPH


The graphs show that people with higher income usually have higher customer lifetime value (CLV), but age also plays a role; older customers  tend to have slightly higher CLV than younger ones. On average, CLV increases a little with age, peaking in the 51–70 group. The strongest drivers of CLV, however, are customer behaviors: buying more often and spending more per purchase both raise CLV, while frequent buyers are much less likely to leave.  In conclusion, it means that this  business should focus on keeping older customers , encouraging all customers to buy more often and spend more, and giving younger high-income customers to grow their long-term value.
